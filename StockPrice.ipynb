# Stock Prediction

This python script will use Long Short Term Memory (LSTM) method to predict the closing stock price using past 60 day stock price. Basically, the data, here stock price of Apple Inc. is taken as example, is scrapped from yahoo and the closed price column is trained and using LSTM model, the prediction is determined. The model looks at past 60 days of data and then gives the prediction accordingly.

# Importing the libraries

```python
import math
import pandas_datareader as web
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

# Reading the data

```python
# Using web.DataReader to grab the data from yahoo.
df = web.DataReader('AAPL', data_source='yahoo', start='2012-01-01', end='2021-05-07')
```

# Visualize the closing price

```python
# Graph of Close Price in USD with respect to date

plt.figure(figsize = (16,8))
plt.plot(df['Close'], linewidth=2)
plt.title('Close price of Apple')
plt.xlabel('Date', fontsize = 18)
plt.ylabel('Close Price in USD($)', fontsize = 18)
plt.show()
```

# Data cleaning

```python
# Creating dataframe with only close price column
data = df.filter(['Close'])

# Converting dataframe into numpy array, so it is easier to do calculations
dataset = data.values

# For training, using 80% of the data and 20% to test the data
trainingDataLen = math.ceil(len(dataset) * 0.8)
```

# Feature Scaling the data

```python
# Using MinMaxScaler to scale the data, which will give the data a number between 0 and 1 accordinly.

scaler = MinMaxScaler(feature_range=(0, 1))
scaledData = scaler.fit_transform(dataset)
scaledData
```

# Training Data Set

```python
trainData = scaledData[0: trainingDataLen, :]

# Splitting the data into xTrain and yTrain
xTrain = []
yTrain = []

# Taking first 60 data points and setting them into as xTrain and 61st data in yTrain
# Again, taking 2nd to 61st data as xTrain and 62nd data in yTrain
# --> So, the model trains with past 60 days and predicts the 61st data
for i in range(60, len(trainData)):
    xTrain.append(trainData[i-60:i, 0])
    yTrain.append(trainData[i, 0])
        
# Converting xTrain and yTrain to numpy arrays
xTrain, yTrain = np.array(xTrain), np.array(yTrain)

# Reshaping the data in 3D, so LSTM can read
xTrain = np.reshape(xTrain, (xTrain.shape[0], xTrain.shape[1], 1))
```

# Building LSTM Model

```python
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(xTrain.shape[1], 1)))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))
```

# Compiling the model

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

# Training the model

```python
model.fit(xTrain, yTrain, batch_size=1, epochs=10)
```

# Creating Test dataset

```python
testData = scaledData[trainingDataLen - 60:, :]

# Creating xTest and yTest
xTest = []
yTest = dataset[trainingDataLen:, :]

for i in range(60, len(testData)):
    xTest.append(testData[i-60:i, 0])
    
# Converting into numpy array
xTest = np.array(xTest)

# Reshaping the data
xTest = np.reshape(xTest, (xTest.shape[0], xTest.shape[1], 1))
```

# Prediction

```python
predictions =  model.predict(xTest)

# Inverse Transform: Unscaling the data
predictions = scaler.inverse_transform(predictions)
```

# Root Mean Squared Error (RMSE)

```python
rmse = np.sqrt(np.mean(predictions - yTest)**2)
rmse
```

# Plotting  the data

```python
train = data[:trainingDataLen]
valid = data[trainingDataLen:]
valid['Predictions'] = predictions

plt.figure(figsize=(16, 8))
plt.plot(train['Close'], linewidth=2)
plt.plot(valid[['Close', 'Predictions']], linewidth=2)
plt.title('Model')
plt.xlabel('Data', fontsize=18)
plt.ylabel('Close Price USD ($)', fontsize=18)
plt.legend(['Train', 'Val', 'Predictions'], loc='lower right')
plt.show()
```
